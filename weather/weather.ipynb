```python
import meteostat
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

start = datetime.now() - timedelta(days=31)
end = datetime.now()

city = meteostat.Point(48.66, -72.52)
data = meteostat.Daily(city, start, end)
data = data.fetch()
data = data.fillna(0)
data[['tavg','tmin', 'tmax', 'prcp', 'snow']]

# average_temperature = data.loc[f'{date.strftime('%Y-%m-%d')}']['tavg']
```

```python
data.plot(y=['tavg', 'tmin', 'tmax'])
plt.title('Temperature Data of The Last 31 Days')
plt.xlabel('Date')
plt.ylabel('temperature')
plt.show()

```

```python
data.plot(y=['prcp'])
plt.title('Total Precipitation')
plt.show()
```

```python
data.plot(y=['snow'])
plt.title('Snow Depth')
plt.show()
```

```python
import folium
import numpy as np
import random
from folium.plugins import HeatMap 
axis = [51.4968, 0.1001]
marks = [[axis[0]+0.20, axis[1]-0.20],[axis[0]-0.0130, axis[1]+0.3012], [axis[0]-0.1530, axis[1]-0.0542], [axis[0]-0.0423, axis[1]-0.2860], [axis[0], axis[1]] ]
x_axis = [mark[0] for mark in marks]
y_axis = [mark[1] for mark in marks]
# we need to find all axises within a certain range and mark them

arr = np.array([axis]*500)

def random_point(axis):
    rand = random.randint(0,3)
    x, y = axis
    if rand == 0:
        x = random.uniform(axis[0], random.choice(x_axis))
    if rand == 1:
        x = random.uniform(axis[0], random.choice(x_axis))
    if rand == 2:
        y = random.uniform(axis[1], random.choice(y_axis))
    if rand == 3:
        y = random.uniform(axis[1], random.choice(y_axis))
    axis[0] = x
    axis[1] = y
    return axis
m = folium.Map(axis, zoom_start=9)
data = []
for i in range(500):
    if i == 0:
        continue
    arr[i] = random_point(arr[i-1])
    data.append(arr[i].tolist())
for i in range(499):
    data[i].append(random.uniform(-12, -11))

HeatMap(data[0:400], min_opacity=0.3, max_val=-11, radius=20).add_to(m)
m
```
